## 🛠️ Understanding Syntax Errors

A **syntax error** happens when Python doesn’t understand the way something is written.  
It’s one of the most common messages beginners see, and it’s completely normal.

Sometimes syntax errors can feel confusing — the message may look long, the wording can be unfamiliar, and it’s not always obvious what went wrong. This is part of learning how Python communicates.

---

### ❓ Why syntax errors happen
Some common causes include:
- Missing quotation marks  
- Forgetting a parenthesis  
- Using a variable name with spaces  
- Mixing single and double quotes in a string  
- Typing something Python doesn’t recognize  

---

### 📍 How Python shows the problem
Python stops running the code and displays an error message.  
The most helpful part is usually near the **bottom**, where it shows:
- the **line number**
- a short explanation of what went wrong

---

### 💡 Tips for interpreting syntax errors
- Read the message slowly — the first glance can feel overwhelming  
- Check the **line number** Python mentions  
- Look for small details: quotes, parentheses, spelling  
- Fix **one** thing at a time  
- If the line looks fine, check the line *above* it  

---

### 🤝 When you need help
If the message still doesn’t make sense, that’s okay — this is a great moment to get support.

You can:
- Ask a classmate or instructor  
- Search the exact error online  
- Use AI (like ChatGPT) to explain the error in plain language  

AI can help you understand what the error means **without taking away the learning**,  
and it can give you simple explanations when you’re stuck.

---

Syntax errors aren’t a sign of failure — they’re part of how Python guides you toward clearer code.

## Understanding Strings in Python

A **string** is a data type in Python used to store text.  
Strings can include letters, numbers, spaces, and punctuation.

You can create a string using **double quotes** or **single quotes** — both work the same:

```python
"California"
'Hello from Oakland'
```

Use whichever feels easier to read.  
The important thing is that the opening and closing quotes match.

### When your string contains quotes
If the text itself includes quotation marks, you have options:

**Use the opposite quote on the outside**
```python
quote = 'The sign said "Welcome to San Jose!"'
```

**Or escape a quote using a backslash `\`**
```python
message = "He said, \"It’s a sunny day in California.\""
```

Both approaches help Python understand where the string begins and ends.

### California Examples
```python
city = "Los Angeles"
state_bird = 'California Quail'
park = "Yosemite National Park"
```

### Why strings matter
- They help your program communicate clearly  
- They’re used for names, places, messages, and labels  
- They make your output easier for real people to understand  

Strings will show up everywhere in your programs — especially when you combine them with input and print statements.

## Understanding Variables in Python

A **variable** is a name you create to store information in your program.  
Python saves the value you assign, and you can use that name later instead of retyping the value.

You create a variable using a **name**, an **equal sign**, and a **value**:

```python
city = "Berkeley"
miles_to_school = 12
temperature = 68.5
```

### How variables work
- The word on the left (`city`) is the **variable name**
- The value on the right (`"Berkeley"`) is what the variable **stores**
- The `=` sign tells Python to **assign** that value to the name

---

## Variable Naming Rules (Beginner-Friendly)

Python gives you a lot of flexibility, but these guidelines help you avoid common errors:

### ✅ Variable names *must*:
- Start with a **letter** or **underscore** (`favorite_food`, `_total`)
- Use only **letters**, **numbers**, and **underscores**
- Have **no spaces**  
  Use underscores instead:  
  `user name` ❌ → `user_name` ✔
- End with anything except punctuation  
  (`score!` ❌)

### ❗ Variable names are **case sensitive**
`City`, `city`, and `CITY` are three different variables in Python.

### ❌ Variable names cannot use Python keywords
These are special words Python already uses, such as:
`if`, `else`, `for`, `while`, `class`, `def`, `return`, etc.

Trying to use one as a variable name will cause an error.

---

## Helpful Naming Tips

These aren’t rules—just good habits:
- Choose names that describe the information  
  (`student_name` instead of `n`)
- Keep names lowercase for readability  
  (`miles_traveled` rather than `MilesTraveled`)
- Be consistent with your style

---

## Why variables matter
Variables let your program:
- Store information once and reuse it
- Work with changing data (like user input)
- Build scripts that make sense to real people

You’ll use variables continuously throughout this module and every module after it.

### Data Types and Literals in Python 📘

Python uses different data types to represent different kinds of information.  
A **literal** is a fixed value that appears directly in your code, such as a number or a piece of text.

Common data types:

- **int**: whole numbers 🔢  
  Example literal: `12`

- **float**: numbers with decimals 💧  
  Example literal: `3.5`

- **str**: text enclosed in quotes 📝  
  Example literal: `"Hello"`

- **bool**: truth values used for logic 🔍  
  Example literals: `True` and `False`

### Non Examples of Literals 🚫

- A variable name such as `total`  
- A function call such as `print("Hello")`  
- An expression such as `x + 5`  
- A placeholder like `input()`

These are not fixed values written directly in your code, so they are not considered literals.

### Understanding the input Function in Python

### What is a Function in Python ❓

A function is a named block of code that performs a specific task.  
You call the function by using its name, the function does its work, and it returns a value or completes an action.  
The `input()` function is one of Python’s built in functions.

---

### Starting with Simple String Input 📥

The easiest way to use `input()` is to collect text from the user.  
Because `input()` always returns a string, collecting words, names, or short messages works smoothly.

Examples:

`name = input("Enter your name: ")`  
`color = input("Enter your favorite color: ")`  

In both examples, whatever the user types is stored as a string.  
Strings require no conversion, so this is a comfortable starting point for new programmers.

---

### Why input Always Gives You a String 🔡

Whenever you use `input()`, Python treats the user’s response as text.  
Even if the user enters a number, Python still stores it as a string.

Example:

`age = input("Enter your age: ")`  

If the user types 25, Python stores `"25"` which is a string, not a number.

---

### Why This Becomes a Problem When You Need Numbers 🔍

Strings cannot be used in math.  
Trying to add a string to a number will cause an error.

Example:

\`\`\`
age = input("Enter your age: ")     # age is "25" as a string
total = age + 5                     # this will fail
\`\`\`

Python cannot add text and a number together.

---

### Converting String Input to Numbers 🔢

When you want to enter a number, you must convert the string to the correct numeric type.

1. Use `int()` for whole numbers  
   Example:  
   `age = int(input("Enter your age: "))`

2. Use `float()` for decimals  
   Example:  
   `price = float(input("Enter a price: "))`

After converting, the value becomes a number that you can safely use in calculations.

(Gaddis 51-52)


````
# Practice with input
# Try entering both strings and numbers

print("Practice Input")

# Collect a string
name = input("Enter your name: ")
print("You entered the name:", name)

# Collect a whole number
age = int(input("Enter your age as a whole number: "))
print("Next year you will be", age + 1)

# Collect a decimal number
price = float(input("Enter a price with decimals: "))
print("The price doubled is", price * 2)
````
